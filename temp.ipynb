```python
#1
def step_fn(z):
    return 1 if z>=0 else 0
def perceptron(x1,x2,w1,w2,b):
    z = (x1*w1)+(x2*w2)+b
    print(f"Net: {z}")
    return step_fn(z)
print(f"Perceptron:{perceptron(1,1,1,1,-0.5)}")
```

```python
#2
import numpy as np
x = np.array([1,1])
w = np.array([1,1])
b = -1.5
net_input = np.dot(x , w) + b
predicted = 1 if net_input>=0 else 0
print(f"net:{net_input},Predicted: {predicted}") 
```

```python
#3
def activations(list_1):
    x = np.arr(list_1)
    step = np.where(x>=0,1,0)
    linear = x
    sigmoid = 1/(1 + np.exp(-x))
    tanh = np.tanh(x)
    relu = np.max(0,x)
    leaky_relu = np.where(x>0,x,x*0.01)
    e_x = np.exp(x - np.max(x))
    softmax = e_x / e_x.sum()

    return ("Step": step,"Linear":linear,"sigmoid":sigmoid,"tanh":tanh,"relu":relu,"Leaky relu": leaky_relu,"softmax":softmax)

y = [-2, 0 ,5, 0, 0, 7, 3]
print(activations(y))
```
